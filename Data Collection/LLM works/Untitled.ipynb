```python
import pandas as pd
import glob
import os

# Folder containing the CSV files
folder_path = r"C:\Users\prart\OneDrive\Desktop\Research_Project\Research Files\Data Collection\LLM works"

# Get all CSV files
csv_files = sorted(glob.glob(os.path.join(folder_path, "*.csv")))

if not csv_files:
    raise FileNotFoundError("No CSV files found in the specified folder.")

# Read all files as strings to avoid data type conversion issues
dataframes = []

for file in csv_files:
    print(f"Reading: {os.path.basename(file)}")
    df = pd.read_csv(file, dtype=str, keep_default_na=False)
    dataframes.append(df)

# Verify that all files have identical columns
reference_columns = list(dataframes[0].columns)

for i, df in enumerate(dataframes[1:], start=2):
    if list(df.columns) != reference_columns:
        raise ValueError(
            f"Column mismatch detected in file #{i}. "
            "All CSV files must have identical columns."
        )

# Merge all rows
merged_df = pd.concat(dataframes, ignore_index=True)

# Save merged file
output_file = os.path.join(folder_path, "merged_output.csv")
merged_df.to_csv(output_file, index=False)

print("\nMerge completed successfully!")
print(f"Files merged: {len(csv_files)}")
print(f"Total rows: {len(merged_df)}")
print(f"Output file: {output_file}")
```
